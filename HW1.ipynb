```python
import numpy as np
import matplotlib.pyplot as plt
```

**Question 1:**

```python
x = np.linspace(0,1,50)
y = np.log(x)
z = np.exp(x)

plt.scatter(x,y, label='Natural Log')
plt.scatter(x,z, label='Exponential')

plt.xlabel('X Input')
plt.ylabel('Function Output')
plt.title('Natural Log and Exponential Function')
plt.legend(loc='lower right')
plt.show()
```

```python
x = np.linspace(-6.5,6.5,100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, label='Sine')
plt.plot(x,z,label='Cosine')

plt.xlabel('X Input')
plt.ylabel('Function Output')
plt.title('Sine and Cosine Functions')
plt.legend(loc='lower left')
plt.show()
```

**Question 4:**

```python
def mathFunction(x):
    y = 100 - 2*x**2 + 3*x
    return(y)

def maxValue(a,b,n,f):
    xValues = np.linspace(a,b,n)
    yValues = f(xValues)
    
    maximum = max(yValues)
    maxIndex = np.argmax(yValues)
    maxXvalue = xValues[maxIndex]
    
    values = {maximum, maxXvalue}
    
    return(values)
    
print(maxValue(-1,1,3,mathFunction))
print(maxValue(-1,1,10,mathFunction))
print(maxValue(-1,1,100,mathFunction))
print(maxValue(-1,1,1000,mathFunction))
print(maxValue(-1,1,5000,mathFunction))
print(maxValue(-1,1,41,mathFunction))
```

The given inputs never return {0.75,101.125} as the max. The true max is at 0.75, but with N = {3, 10, 100, 1000, 5000} in [0,1], 0.75 is never a value in any of these input arrays. Therefore, the true maximum is never reached. If N = 41, for example, then 0.75 is part of the given inputs and thus the theoretical maximum is returned.

However, just like a limit (because it is one), increasing the value of N increases the accuracy of the maximum function value since it is approaching the true maximum input of 0.75.
